# **Replication Data Preprocessing**

```python
#preprocess and rescale data
import scipy.io
import pandas as pd
import numpy as np

dict = {'id':[], 'happiness rating':[], 'certain reward':[], 'expected value':[], 'prediction error':[]}

# get data for each participant
for k in range(111, 161):
    file_name = '../Data/Raw Data/test'+str(k)+'.mat'

    data_mat = scipy.io.loadmat(file_name, simplify_cells=True)

    behavedata = pd.DataFrame(data_mat["data"]["behavedata"]).iloc[:, :12]
        
    behavedata.columns = [
            "trial_number", "gambleside", "certain", "win", "loss", "buttonpress", 
            "chosegamble", "outcome", "choice_RT", "happiness_rating", 
            "happiness_rating_startingpoint", "happiness_rating_RT"
        ]

    print(behavedata.tail(50))

    # get data from each trial
    for i in range(1, 151):
        if(behavedata['chosegamble'][i] == 0.0):
            dict["certain reward"].append(behavedata["certain"][i])
            dict["expected value"].append(0)
            dict["prediction error"].append(0)
        else:
            dict["certain reward"].append(0)
            EV = (behavedata["win"][i]+behavedata["loss"][i])/2
            dict["expected value"].append(EV)
            dict["prediction error"].append(behavedata["outcome"][i]-EV) 
        dict["id"].append(k)
        dict["happiness rating"].append(behavedata["happiness_rating"][i]*100)
df = pd.DataFrame(dict)

# remove incomplete time series
df = df[df['id']!=151]
df.to_csv("../Data/Replication_data_rescale.csv")    
```
